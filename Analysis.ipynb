```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_complete = "PyCitySchools/Resources/schools_complete.csv"
student_complete = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_complete = pd.read_csv(school_complete)
student_complete = pd.read_csv(student_complete)

# Combine the data into a single dataset.  
school_student_complete = pd.merge( student_complete,school_complete, how="left", on=["school_name", "school_name"])
school_student_complete
```

```python
#Total schools
total_schools =school_student_complete["school_name"].nunique()
total_schools
```

```python
#Total students
total_students = school_student_complete["Student ID"].nunique()
total_students
```

```python
#Total Budget
total_budget = school_complete["budget"].unique()
budget = total_budget.sum()
budget
```

```python
#Average Math Score
average_math_score = school_student_complete["math_score"].mean()
average_math_score 
```

```python
#Average Reading Score
average_reading_score = school_student_complete["reading_score"].mean()
average_reading_score
```

```python
# % Passing Math 
passing_math = len(school_student_complete[school_student_complete["math_score"] >= 70])
math_percent = passing_math / total_students * 100
math_percent
```

```python
# % Passing Reading
passing_reading = len(school_student_complete[school_student_complete["reading_score"] >=70])
reading_percent = passing_reading / total_students * 100
reading_percent
```

```python
# % Overall Passing
overall_passing = school_student_complete[(school_student_complete["math_score"] >=70) & (school_student_complete["reading_score"] >=70)].count()["student_name"]
overallpassing_percent = overall_passing / (total_students) * 100
overallpassing_percent
```

```python
#Distrcit Summary Table 
table_df = pd.DataFrame({"Total School": [total_schools], "Total Students": [total_students], "Total Budget": [budget], "Average Math Score": [average_math_score], "Average Reading Score": [average_reading_score], "% Passing Math": [math_percent], "% Passing Reading": [reading_percent], "% Overall Passing": [overallpassing_percent]})
table_df
```

# Total School Budget

```python
#Total School BUDGET
school_budget_total = school_complete.groupby(["school_name"])["budget"].mean()
school_budget_total
```

# Budget per Student 

```python
#Student Budget per Student
per_stud_budget = school_budget_total /students_per_school
per_stud_budget
```

# Average Math Score per School

```python
#Average math score per school
average_math = school_student_complete.groupby(["school_name"])["math_score"].mean()
average_math
```

# Average Reading Scores

```python
average_reading = school_student_complete.groupby(["school_name"])["reading_score"].mean()
average_reading
```

```python
#PERCENT of Math Passing Per School
passing_math = school_student_complete[school_student_complete["math_score"] >= 70]

per_of_math = passing_math.groupby(["school_name"]).count()["student_name"] / students_per_school *100
per_of_math.head()

```

```python
#PERCENT of Reading Per School
passing_reading = school_student_complete[school_student_complete["reading_score"] >=70]


per_of_read = passing_reading.groupby(["school_name"]).count()["student_name"] / students_per_school *100
per_of_read.head()
```

```python
#OVERALL Passing Rate (Average)

overall = per_of_read +per_of_math / students_per_school
overall
```

# School Summary 

```python
#New DataFrame for School Summary

school_summary =({
    "School Type": school_type,
    "Total Students":students_per_school,
    "Total School Budget": total_budget,
    "Per Student Budget": per_stud_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": per_of_math,
    "% Passing Reading": per_of_read,
    "% Overall Passing": overall})

school_summary =pd.DataFrame(school_summary)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary
```
